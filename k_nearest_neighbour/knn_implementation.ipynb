```python
import k_nearest_neighbour as k
import numpy as np
dat = np.loadtxt(open("/content/knn_data.txt"),delimiter=',')
np.random.shuffle(dat)
x = dat[:,0:-1]
y = dat[:,-1:]
(a,b)=np.shape(x)
dp = k.knn()
x = dp.normalize(x)
x_train=x[0:7*a//10,:]
y_train=y[0:7*a//10,:]
x_test=x[7*a//10:a,:]
y_test=y[7*a//10:a,:]
```

```python
dp.fit(x_train,y_train,4)
print("Some predicted values:\n",dp.predict(x_test).flatten()[0:10])
print("Corresponding actual values:\n",y_test.flatten()[0:10])
print("Accuracy of my model:",dp.accuracy(x_test,y_test),"%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
ap=KNeighborsClassifier(4)
ap.fit(x_train,y_train.flatten())
print("Some predictions of sklearn:\n",ap.predict(x_test).flatten()[0:10])
print("Corresponding actual values:\n",y_test.flatten()[0:10])
print("Accuracy of sklearn:",np.mean(y_test.flatten()==ap.predict(x_test).flatten())*100,"%")
```
